# Visualizations

```python
# Gender distribution with the help of pie chart.

gender_count = Data[' Gender'].value_counts()
gender_count
```

```python
plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_count, labels = gender_count.index, autopct= '%1.1f%%', startangle = 150, shadow = True);
```

```python
# Age distribution of the of employees.

plt.figure(figsize=(12,6))
plt.title('Age Distribution of the Employees')
plt.xlabel(' Age')
plt.ylabel('Number of Employees')
plt.hist(Data[' Age'], bins = np.arange(10,80,2), color='orange', edgecolor = 'white');
```

```python
#  Now Checking type of Occupattion based on Genders

male = Data[Data[' Gender'] == 'Male']
female = Data[Data[' Gender'] == 'Female']
```

```python
plt.figure(figsize=(12,8))
sns.countplot(y = ' Occupation' , order = Data[' Occupation'].value_counts().index[0:10], data = male)
plt.title('Top 10 Occupations of Male')




plt.figure(figsize=(12,8))
sns.countplot(y = ' Occupation' , order = Data[' Occupation'].value_counts().index[0:10], data = female)
plt.title('Top 10 Occupations of Female ')
```

```python
# checking what number of age has the highest salary as per the Gender.

plt.figure(figsize=(12,10))
axis = sns.scatterplot(x = ' Age', y = ' Salary',  data = Data, hue = ' Gender')
plt.title('Highest Salary with respect to Age and Gender')
```

```python
# In Occupation how much salary employee have as per there age.

plt.figure(figsize =(20,18))
sns.heatmap(Data.pivot_table(index = Data[' Occupation'] , columns = ' Age', values = ' Salary', aggfunc='count').fillna(0).astype(int),annot = True)
```

```python
# Distribution of Children with respect to Employees Age

fig = plt.figure(figsize = (16,8))
sns.boxplot(x = Data[' No. of Children'], y = Data[' Age'], data = Data)
plt.ylim(0, 100)
plt.title('Distribution of children With Resspect to Employees Age')
```

```python
# how many employee work in each Occupation.

plt.figure(figsize=(15,15))
sns.countplot( y = ' Occupation' , data = Data)
plt.title('occupation cout')
```
